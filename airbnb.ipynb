# kaggle airbnb feature study

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Draw inline
%matplotlib inline

# Set figure aesthetics
sns.set_style("white", {'ytick.major.size': 10.0})
sns.set_context("poster", font_scale=1.1)
```

```python
train_users = pd.read_csv("./input/sessions.csv")
```

```python
train_users.head(11)
```

```python
users_action = train_users['action']
users_action_type = train_users['action_type']
users_action_detail = train_users['action_detail']
users_device_type = train_users['device_type']
users_action.describe()
```

```python
train_users['action'].count()
```

```python
plt.figure(figsize=(30,10))
users_action_type.value_counts(dropna=False).plot(kind='bar', color='#FD5C64', rot=0)
plt.xticks(rotation=40, ha='center')
plt.xlabel('users_action_type')
sns.despine()
```

```python
plt.figure(figsize=(15,8))
users_action_type.value_counts(dropna=False).plot(kind='bar', color='#FD5C64', rot=0)
plt.xticks(rotation=40, ha='center')
plt.xlabel('action')
sns.despine()

```
